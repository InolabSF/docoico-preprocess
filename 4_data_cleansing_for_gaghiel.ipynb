```python
import pandas as pd
import numpy as np
import os
import urllib.request
import json

pd.set_option('display.max_columns', 200)
```

```python
#destination
df = pd.read_csv('./boosted/destination.csv', sep=',')
columns = df.columns.drop(['Address', 'Image', 'Subtitle', 'Desc', 'Category', 'Created at', 'Updated at', 'Id [Destination articles]', 'Destination article category [Destination articles]', 'Title [Destination articles]', 'Desc [Destination articles]', 'Image [Destination articles]', 'Created at [Destination articles]', 'Updated at [Destination articles]'])
df_destination = df[columns]

columns = ['Food & Drink', 'Museums', 'Nature & Parks', 'Sights & Landmarks', 'Zoos & Aquarium']
for column in columns:
    df_destination[column] = df_destination[column].fillna(0).astype(int)

df_destination.to_csv('./gaghiel/destination.csv', encoding='utf-8', index=False)
df_destination
```

```python
# user
df = pd.read_csv('./boosted/user.csv', sep=',')
columns = df.columns.drop(['Email', 'Reset password sent at', 'Remember created at', 'Sign in count', 'Current sign in at', 'Last sign in at', 'Current sign in ip', 'Last sign in ip', 'Created at', 'Updated at', 'Admin', 'Line', 'Name', 'Access token', 'Refresh token'])
df_user = df.query('Admin != "t"')[columns]
df_user = df_user.rename(columns={'Id': 'User'})

# user_destination
df = pd.read_csv('./tables/user_destination.csv', sep=',')
columns = df.columns.drop(['Id', 'User result', 'Value', 'Created at', 'Updated at'])
df_user_destination = df.query('Value > 0')[columns]

# destination
df_destination = pd.read_csv('./boosted/destination.csv', sep=',')


series = { 'Destination': [], 'TripAdvisor': [], 'Prefecture': [], 'Settlement': [], 'Gcloud detect-labels': []}
tripadvisor_columns = ['Food & Drink', 'Museums', 'Nature & Parks', 'Sights & Landmarks', 'Zoos & Aquarium']
prefecture_columns = list(set(df_destination['administrative_area_level_1'].values))
settlement_columns = ['urban', 'rural']
gcloud_columns = []
for column in df_destination.columns:
    if column.startswith('Gcloud detect-labels ['):
        gcloud_columns.append(column.replace('Gcloud detect-labels [', '').replace(']', ''))
        
        
for i, user_row in df_user.iterrows():
    df = df_user_destination.loc[df_user_destination['User'] == user_row['User']]
    
    tripadvisors = {}
    for tripadvisor_column in tripadvisor_columns:
        tripadvisors[tripadvisor_column] = 0
    prefectures = {}
    for prefecture_column in prefecture_columns:
        prefectures[prefecture_column] = 0
    settlements = {}
    for settlement_column in settlement_columns:
        settlements[settlement_column] = 0
    gclouds = {}
    for gcloud_column in gcloud_columns:
        gclouds[gcloud_column] = 0    
        
    destinations = []
    if not df.empty:
        destinations = df['Destination'].values
    destination = ""
    for j, destination_row in df_destination.iterrows():
        new_destination = ('%d: 0;' % destination_row['Id'])
        if destination_row['Id'] in destinations:
            new_destination = ('%d: 1;' % destination_row['Id'])
            
            for tripadvisor_column in tripadvisor_columns:
                if destination_row[tripadvisor_column] > 0:
                    tripadvisors[tripadvisor_column] = tripadvisors[tripadvisor_column] + 1
            
            for prefecture_column in prefecture_columns:
                if destination_row['administrative_area_level_1'] == prefecture_column:
                    prefectures[prefecture_column] = prefectures[prefecture_column] + 1
            
            if isinstance(destination_row['administrative_area_level_2'], str):
                settlements['rural'] = settlements['rural'] + 1
            else:
                settlements['urban'] = settlements['urban'] + 1
                
            for gcloud_column in gcloud_columns:
                if destination_row["Gcloud detect-labels [%s]" % gcloud_column] > 0:
                    gclouds[gcloud_column] = gclouds[gcloud_column] + 1
                    
        destination = destination + new_destination
    
    tripadvisor = ""
    for key, value in tripadvisors.items():
        tripadvisor = tripadvisor + "%s: %s;" % (key, value)
    prefecture = ""
    for key, value in prefectures.items():
        prefecture = prefecture + "%s: %s;" % (key, value)
    settlement = ""
    for key, value in settlements.items():
        settlement = settlement + "%s: %s;" % (key, value)
    gcloud = ""
    for key, value in gclouds.items():
        gcloud = gcloud + "%s: %s;" % (key, value)
        
    series['Destination'].append(destination)
    series['TripAdvisor'].append(tripadvisor)
    series['Prefecture'].append(prefecture)
    series['Settlement'].append(settlement)
    series['Gcloud detect-labels'].append(gcloud)

    
df_user['Destination'] = pd.Series(series['Destination']).values
df_user['TripAdvisor'] = pd.Series(series['TripAdvisor']).values
df_user['Prefecture'] = pd.Series(series['Prefecture']).values
df_user['Settlement'] = pd.Series(series['Settlement']).values
df_user['Gcloud detect-labels'] = pd.Series(series['Gcloud detect-labels']).values



df_user.to_csv('./gaghiel/user.csv', encoding='utf-8', index=False)
df_user
```
