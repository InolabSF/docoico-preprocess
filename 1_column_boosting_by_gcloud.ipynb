````python
import numpy as np
import pandas as pd
import os
import urllib.request
import urllib.parse
import json

GOOGLE_API_KEY = os.environ['GOOGLE_API_KEY']
pd.set_option('display.max_columns', 200)
df = pd.read_csv('./tables/destination.csv', sep=',')
df
````

````python
# expand columns by lat lng
df = pd.read_csv('./tables/destination.csv', sep=',')
addresses = []
for index, row in df.iterrows():
    list = {}
    with urllib.request.urlopen("https://maps.googleapis.com/maps/api/geocode/json?latlng=%s,%s&key=%s" % (row['Lat'], row['Lng'], GOOGLE_API_KEY)) as url:
        data = json.loads(url.read().decode())
        if data['status'] != 'OK':
            continue
        address_components = data['results'][0]['address_components']
        for address_component in address_components:
            types = address_component['types']
            for type in types:
                list[type] = address_component['long_name']
        addresses.append(list)
lists = {}
for address in addresses:
    for key, value in address.items():
        if (key in lists) == False:
            lists[key] = []
for address in addresses:
    for key, value in lists.items():
        if key in address:
            value.append(address[key])
        else:
            value.append('')
for key, value in lists.items():
    df[key] = pd.Series(value).values
df.to_csv('./boosted/destination.csv', encoding='utf-8', index=False)
df
````

````python
lists = []
mids = []
df = pd.read_csv('./boosted/destination.csv', sep=',')
for index, row in df.iterrows():
    mid = {}
    command = "gcloud ml vision detect-labels %s" % row['Image']
    try:
        response = os.popen(command).read()
        data = json.loads(response)
        for response in data['responses']:
            for labelAnnotation in response['labelAnnotations']:
                desc = labelAnnotation['mid']
                mid[desc] = labelAnnotation['score']
                if (desc in lists) == False:
                    lists.append(desc)
    except:
        print("Failed: %s" % command)
    mids.append(mid)
for mid in mids:
    for key in lists:
        if (key in mid) == False:
            mid[key] = 0.0
fileds = lists
lists = {}
for filed in fileds:
    lists[filed] = []
for mid in mids:
    for key, value in mid.items():
        lists[key].append(value)
for key, value in lists.items():
    df["Gcloud detect-labels [%s]" % key] = pd.Series(value).values
df.to_csv('./boosted/destination.csv', encoding='utf-8', index=False)
df
````

````python
df = pd.read_csv('./boosted/destination.csv', sep=',')
mids = []
for column in df.columns:
    if column.startswith('Gcloud detect-labels ['):
        mids.append(column.replace('Gcloud detect-labels [', '').replace(']', ''))


types = []
names = []
for index, mid in enumerate(mids):
    with urllib.request.urlopen("https://kgsearch.googleapis.com/v1/entities:search?ids=%s&key=%s" % (mid, GOOGLE_API_KEY)) as url:
        data = json.loads(url.read().decode())
        t = []
        n = []
        if 'itemListElement' in data:
            for itemListElement in data['itemListElement']:
                if 'name' in itemListElement:
                    n.append(itemListElement['name'])
                if 'result' in itemListElement:
                    if '@type' in itemListElement['result']:
                        t = t + itemListElement['result']['@type']
        types.append(set(t))
        names.append(set(n))

d = {'mid': mids, 'type': types, 'name': names}
df = pd.DataFrame(data=d)
df.to_csv('./boosted/detect_label.csv', encoding='utf-8', index=False)
df
````
